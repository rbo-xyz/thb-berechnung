```python
import pandas as pd

# Beispiel-DataFrame
df = pd.DataFrame({
    "Name": ["Anna", "Bernd", "Claudia"],
    "Alter": [23, 35, 29],
    "Stadt": ["Berlin", "Hamburg", "Muenchen"]
})

# Breiten und Ausrichtung definieren
# 'l' = linksbündig, 'r' = rechtsbündig, 'c' = zentriert
col_format = {
    "Name": (20, 'l'),
    "Alter": (6, 'r'),
    "Stadt": (15, 'c')
}

# Funktion, um eine Zeile zu formatieren
def format_row(row, col_format):
    cells = []
    for col, (width, align) in col_format.items():
        val = str(row[col])
        if align == 'l':
            val = val.ljust(width)
        elif align == 'r':
            val = val.rjust(width)
        elif align == 'c':
            val = val.center(width)
        cells.append(val)
    return "| " + " | ".join(cells) + " |"

# Separator bauen
sep = "+-" + "-+-".join("-" * width for width, _ in col_format.values()) + "-+"

# Tabelle zusammenstellen
lines = [sep, format_row({col: col for col in df.columns}, col_format), sep]
for _, row in df.iterrows():
    lines.append(format_row(row, col_format))
    lines.append(sep)

table_str = "\n".join(lines)

# Ausgabe
print(table_str)

# Optional: in Datei speichern
with open("output.txt", "w", encoding="utf-8") as f:
    f.write(table_str)
```

```python
import pandas as pd

# Beispiel-DataFrame mit Fließkommazahlen
df = pd.DataFrame({
    "Name": ["Anna", "Bernd", "Claudia"],
    "Alter": [23, 35, 29],
    "Note": [3.5, 2.75, 4.2]
})

# Breiten und Ausrichtung: 'l', 'r', 'c', 'p' = Punktbündig
col_format = {
    "Name": (12, 'l'),
    "Alter": (6, 'r'),
    "Note": (6, 'p')  # Punktbündig
}

def format_float_column(series, width):
    """Fließkommazahlen punktbündig formatieren"""
    # Strings mit 2 Nachkommastellen
    strs = series.map(lambda x: f"{x:.2f}")
    # Max Länge links und rechts vom Punkt
    left_lengths = strs.map(lambda x: len(x.split('.')[0]))
    right_lengths = strs.map(lambda x: len(x.split('.')[1]))
    max_left = left_lengths.max()
    max_right = right_lengths.max()
    
    # Füllen links und rechts
    return strs.map(lambda x: x.split('.')[0].rjust(max_left) + '.' + x.split('.')[1].ljust(max_right))

def format_row(row, df, col_format):
    cells = []
    for col, (width, align) in col_format.items():
        val = row[col]
        if align == 'l':
            val = str(val).ljust(width)
        elif align == 'r':
            val = str(val).rjust(width)
        elif align == 'c':
            val = str(val).center(width)
        elif align == 'p':
            val = str(val).rjust(width)  # Punktbündig schon vorbereitet
        cells.append(val)
    return "| " + " | ".join(cells) + " |"

# Punktbündige Spalten vorbereiten
for col, (width, align) in col_format.items():
    if align == 'p':
        df[col] = format_float_column(df[col], width)

# Separator bauen
sep = "+-" + "-+-".join("-" * width for width, _ in col_format.values()) + "-+"

# Tabelle zusammenstellen
lines = [sep, format_row({col: col for col in df.columns}, df, col_format), sep]
for _, row in df.iterrows():
    lines.append(format_row(row, df, col_format))
    lines.append(sep)

table_str = "\n".join(lines)
print(table_str)
```
